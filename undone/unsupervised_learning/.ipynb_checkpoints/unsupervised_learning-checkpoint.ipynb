В обучении с учителем у нас есть признаки $X$ и целевая переменная $y$, которая является "учителем" для модели. Алгоритм использует правильные ответы, чтобы понимать, насколько верны его предсказания.

В обучении без учителя у нас есть только признаки $X$. Модель учится самостоятельно обнаруживать закономерности и структуру в данных. Оценивать качество работы такого алгоритма достаточно сложно, так как нет правильных ответов.

##### Неразмеченные данные

Данные – это очень ценный ресурс, который является залогом успеха многих ML проектов. Хорошие данные могут быть важнее самых новых и сложных моделей.

Получить размеченные данные может быть очень сложно и дорого, особенно если данные специфичные (например, медицинские данные) или если речь идёт про изображения, видео, тексты и звук. 

Неразмеченных данных гораздо больше, чем размеченных. Табличные данные, видео, изображения, тексты и т.д., которые находятся в открытом доступе – всё это огромный потенциал для построения моделей. Сейчас активно разрабатываются подходы по работе с неразмеченными данными, особенно в сфере глубокого обучения, где нейронные сети требуют больших объёмов данных. Например, в [этой статье](https://arxiv.org/pdf/2007.11984.pdf) авторы используют соседние кадры видео в качестве целевой переменной.

Из неразмеченных данных можно извлекать много полезной информации. Мы можем понять структуру наших данных, разделить объекты на группы с похожими характеристиками. Мы также можем использовать неразмеченные данные для того, чтобы сгенерировать дополнительные признаки для нашей основной модели.

**Мягкая кластеризация** – для каждого объекта определяется вероятность принадлежности к каждому кластеру.

### K-Means

K-means – алгоритм кластеризации, который минимизирует расстояния от объектов до центров внутри каждого кластера.

$$J = \sum_{k=1}^K \sum_{x^{(i)} \in k}(x^{(i)} - \mu_k)^2$$

Алгоритм работает итеративно, на каждом шаге уменьшая функцию ошибки $J$ – сумму квадратов расстояний между объектами и центрами кластеров, к которым они принадлежат. Центры кластеров также называются центроидами и обозначаются как $\mu_1, \mu_2, ..., \mu_k$, где $K$ – это количество кластеров.

#### Алгоритм K-means:

- Выбрать количество кластеров K, на которые будут кластеризованы данные;
- Инициализировать положение центроидов кластеров;
- Каждый объект присвоить ближайшему центроиду;
- Пересчитать координаты центроидов как среднее координат отнесённых к ним объектов;
- Повторять 3 и 4, пока не достигнем максимального числа итераций или пока положения центроидов не перестанут меняться.

Разберёмся, на каком шаге и как алгоритм оптимизирует функцию ошибки:

- На шаге 3 координаты центроидов фиксированы, и мы определяем, к какому кластеру отнести объект. Координаты объектов также фиксированы, поэтому определяем объект $x^{(i)}$ в тот кластер, для которого расстояние от объекта до его центроида минимально, то есть $arg min_k (x^{(i)} - \mu_k)^2$
- На шаге 4 мы находим координаты центроидов $\mu_1, \mu_2, ..., \mu_k$, как константы, которые минимизируют функцию ошибки $J$ при фиксированной принадлежности объектов к кластерам. В случае K-means $J$ – это среднеквадратичная ошибка, поэтому центроиды рассчитываются как среднее координат объектов. В модификации K-medians в качестве $J$ выступает средняя абсолютная ошибка, поэтому центроиды рассчитываются как медиана координат объектов.

#### Особенности K-means

1. Необходимо задавать количество кластеров.

Это может быть недостатком алгоритма, если мы не понимаем структуру данных и не знаем, на сколько групп мы хотим разделить объекты. В этом случае, мы должны подбирать количество кластеров $K$ как гиперпараметр алгоритма.

```python
Image('unsupervised_learning/photos/6.png')
```

2. Разбиение на кластеры существенно зависит от инициализации центроидов.

Если неудачно задать начальное положение центров кластеров, то алгоритм может не сойтись в глобальный оптимум.

```python
Image('unsupervised_learning/photos/7.png')
```

3. Необходимо нормировать данные перед подачей в алгоритм.

Разные по масштабу признаки могут приводить к неправильной кластеризации из-за того, что больший по масштабу признак будет вносить больший вклад в функцию ошибки.

```python
Image('unsupervised_learning/photos/8.png')
```

4. K-means хорошо работает на данных, где выделяются сферические кластеры с одинаковой дисперсией.

Все объекты, находящиеся ближе к центроиду, чем самый дальний объект кластера, также должны находиться в этом кластере. Поэтому образуются кластеры сферической формы с радиусом, равным расстоянию от центроида до самого дальнего объекта кластера. 

```python
Image('unsupervised_learning/photos/9.png')
```

5. Гарантированная сходимость за конечное число итераций.

K-means может не сойтись в глобальный минимум, но локальный минимум достигнет в любом случае, поскольку множество объектов конечно, а значение ошибки уменьшается на каждой итерации алгоритма.
 
6. Один объект – один кластер.

K-means – алгоритм жёсткой кластеризации, то есть каждый объект будет отнесён только к одному кластеру. Существует модификация [Fuzzy c-means](http://web-ext.u-aizu.ac.jp/course/bmclass/documents/FCM%20-%20The%20Fuzzy%20c-Means%20Clustering%20Algorithm.pdf), работающая по принципу K-means, но возвращающая вероятности принадлежности к кластерам.
 
7. Неустойчивость к выбросам.

Центроиды могут смещаться в сторону выбросов, чтобы уменьшить значение функции ошибки.

```python
Image('unsupervised_learning/photos/10.png')
```

8. Возможность дообучать алгоритм на новых данных.

При добавлении новых данных K-means запустит несколько дополнительных итераций: скорректирует положение центроидов и определит новые объекты в кластеры.

#### Варианты инициализации

K-means чувствителен к выбору начального положения центроидов. Инициализировать положения центроидов можно так:

1. Случайно выбрать $K$ точек из исходной выборки
2. [kmeans++](http://ilpubs.stanford.edu:8090/778/1/2006-13.pdf) — первый центроид инициализируется случайной точкой из исходной выборки, а все последующие $K−1$ центроид размещаются как можно дальше от предыдущих. В качестве следующего центроида каждая точка выбирается c вероятностью $p_x = \frac{d_x^2}{\sum_{x \in X}{d_x^2}}$, где $d_x$ — расстояние от точки $x$ до ближайшего центроида.


#### Выбор количества кластеров

Качество кластеризации K-means сильно зависит от выбранного количества кластеров $K$. При выборе можно опираться на бизнес логику решаемой задачи или следить за значениями различных метрик:

1. Опора на бизнес смысл задачи

Допустим, мы решаем задачу кластеризации текстов объявлений и хотим выделять 4 состояния качества объекта продажи: "Новое", "Отличное", "Хорошее", "Плохое". Из поставленной нами задачи понятно, что $K=4$.
 
2. "Метод локтя" (Elbow method)

Перебираем различные $K$ и для каждого варианта кластеризации считаем значение функции ошибки: 

$$J = \sum_{k=1}^K \sum_{x^{(i)} \in k}(x^{(i)} - \mu_k)^2$$

```python
Image('unsupervised_learning/photos/11.gif')
```

Данная функция ошибки достигает своего минимума в нуле, когда $K=n$, то есть когда каждый объект является отдельным кластером. Однако такая кластеризация не имеет никакого смысла.
Поэтому выберем такое $K$, начиная с которого ошибка начинает уменьшаться не так сильно.

```python
Image('unsupervised_learning/photos/12.png')
```

3. По метрикам качества кластеризации

Перебираем различные $K$, оцениваем метрики качества каждого варианта кластеризации и выбираем $K$, при котором значение метрик максимально. О том, какие существуют метрики качества кластеризации, мы поговорим в следующих шагах.
 
4. По метрике качестве финальной задачи

Если результат кластеризации используется как признак в другой модели, то необходимо выбрать такое $K$, при котором улучшение метрики качества итоговой модели будет максимально.

#### Задача анализа заголовков в категории «Товары для школы»

Цель — проанализировать, какие товары продаются в данной категории и какие заголовки там бывают для последующего использования этой информации при решении уже какой-то конкретной задачи.

Мы имеем дело с текстовыми данными — чтобы кластеризовать их, нам необходимы вектора. Одними из самых простых методов получения векторов, пригодных для использования в реальных задачах, являются TfidfVectorizer или Word2Vec.

В этом шаге мы воспользуемся моделью Word2Vec, предполагая, что она лучше подходит для смыслового анализа слов.

```python
data = pd.read_csv('titles_for_clustering_original.csv')
```

```python
data.head(3)
```

```python
data.info()
```

```python
# предобработка
def preprocessing(text):
    """Делаем предобработку.
    
    - приводим к нижнему регистру
    - удаляем пунктуацию и цифры
    - удаляем английские слова: не имеют большой значимости для данной задачи
    :return: строка обработанного текста
    """
    text = re.sub(r'[^А-я]', ' ', text.lower())
    return ' '.join(text.split())

data['title'] = data['title'].apply(preprocessing)
# удаляем заголовки, которые стали пустыми после предобработки
mask = data['title'].apply(lambda x: len(x.split()) > 0)
data = data[mask]

# data.drop_duplicates('title', inplace=True) - опционально, зависит от цели кластеризации
data = data.reset_index(drop=True)
```

```python
data.head(3)
```

```python
# обучение Word2Vec
w2v_corpus = [x.split() for x in data['title']]

w2v_model = Word2Vec(min_count=0, vector_size=100, hs=1, window=1, seed=0)
w2v_model.build_vocab(w2v_corpus)
w2v_model.train(w2v_corpus, total_examples=w2v_model.corpus_count, epochs=w2v_model.epochs) 
```

```python
w2v_model.wv['учебник']
```

```python
np.linalg.norm(w2v_model.wv['учебник'] - w2v_model.wv['книга'])
```

```python
# получение векторов заголовков из векторов слов
w2v_embeddings = []
for title_words in w2v_corpus:
    # не забываем, что некоторые слова могут не быть в словаре Word2Vec из-за выставленного min_count
    title_word_embs = [
        w2v_model.wv[word] 
        for word in title_words 
        if word in w2v_model.wv.key_to_index
    ]
    title_emb = sum(title_word_embs) / len(title_word_embs)
    w2v_embeddings.append(title_emb)
w2v_embeddings = np.vstack(w2v_embeddings)
```

#### Реализация KMeans

В этом задании требуется реализовать класс Kmeans , используя шаблон. 

Алгоритм Kmeans:

- Выбрать количество кластеров n_clusters, на которые будут кластеризованы данные;
- Инициализировать центроиды кластеров;
- Каждую точку присвоить ближайшему центроиду;
- Пересчитать координаты центроидов как среднее координат относящихся к ним точек;

- random — случайным образом выбрать n_clusters точек выборки

k-means++  — представляет собой попытку отодвинуть начальные центроиды как можно дальше друг от друга, покрывая как можно больше пространство данных, тем самым способствует устранению проблем при неудачной случайной инициализации. Оригинальная статья по [kmeans++](http://ilpubs.stanford.edu:8090/778/1/2006-13.pdf).

##### Инициализация k-means++:

- В качестве первого центроида $c_1$ выбрать случайную точку из выборки X;
- Для каждой точки $x \in X$ вычислить вероятность быть выбранной в качестве следующего центроида $p_x = \frac{d_x^2}{\sum_{x \in X}{d_x^2}}$, где $d_x$ — расстояние от точки $x$ до ближайшего центроида;
- В качестве следующего центроида $c_i$ выбрать точку $x \in X$ c вероятностью $p_x$;
- Повторять 2 и 3 до тех пор, пока не будет выбрано необходимое количество центроидов — n_clusters.

##### Формат входных данных

В первой строчке $n_1, n_2$ — размеры обучающей и тестовой выборки соответственно и k — число кластеров. Следующие $n_1$ строк - вектора обучающей выборки, за ними $n_2$ строк — вектора тестовой выборки.

В качестве ответа нужно вернуть метки кластеров: в первой строке метки на обучающей выборке, во второй — на тестовой.

#### Визуализация кластеров

```python
import random

import matplotlib.cm as cm
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import euclidean_distances
```

Часто используются два базовых метода анализа получившихся кластеров:

- Посмотреть на несколько ближайших объектов до каждого центроида, исходя из этого оценить содержимое всего кластера.

```python
kmeans = KMeans(n_clusters=19)
kmeans.fit(w2v_embeddings)

top = 5
clusters = kmeans.labels_
dist_to_centers = euclidean_distances(kmeans.cluster_centers_, w2v_embeddings)
top_most_similar = np.argsort(dist_to_centers)[:, :top]

for clt in np.unique(clusters):
    ids = top_most_similar[clt]
    most_similar_titles = '\n'.join(data.iloc[ids]['title'])
    print(f'Cluster {clt}:\n{most_similar_titles}')
```

- Визуализация кластеров в 2D с помощью TSNE — широко распространённый метод визуализации данных высокой размерности. В случае очень большой размерности признакового пространства имеет смысл предварительно уменьшить размерность, например, с помощью PCA, для ускорения вычислений и устранения возможного шума.

```python
def tsne_plot(embeddings, clusters, texts, not_annotate_thr=0.3):
    '''Рисует вектора в 2D, используя TSNE.
    
    :param embedding: вектора заголовков, которые мы хотим нарисовать 
    :param clusters: предсказанные метки кластеров
    :param texts: заголовки 
    :param not_annotate_thr: доля точек, которые не будут подписаны текстом
    '''
    # для каждого кластера сэмплируем не более 20 точек для ускорения TSNE и наглядности картинки
    sample_i = []
    for clt in np.unique(clusters):
        cluster_mask = clusters == clt
        sample_i += random.sample(list(np.where(cluster_mask)[0]), min(20, cluster_mask.sum()))
        
    # берем слайс по sample_i
    embeddings = embeddings[sample_i]
    clusters = clusters[sample_i].reshape(-1, 1)
    texts = texts[sample_i]
    
    # обучаем и применяем TSNE
    tsne_model = TSNE(perplexity=10, metric='cosine', init='pca', n_iter=2500, random_state=17)
    reduced_embeddings = tsne_model.fit_transform(embeddings)
    
    plt.figure(figsize=(18, 18)) 
    colors = cm.nipy_spectral(clusters.astype(float) / len(np.unique(clusters)))
    for i, (x, y) in enumerate(reduced_embeddings):
        if np.random.rand() > not_annotate_thr:
            plt.scatter(x, y, c=colors[i])
            plt.annotate(texts[i],
                         xy=(x, y),
                         xytext=(5, 2),
                         textcoords='offset points',
                         ha='right',
                         va='bottom', 
                         size=8)
    plt.show()
    
tsne_plot(w2v_embeddings, clusters, data['title'].values)   
```

### DBSCAN: краткий обзор

Одним из наиболее часто используемых на практике методов кластеризации при неизвестном количестве кластеров является [DBSCAN](https://ru.wikipedia.org/wiki/DBSCAN) — метод кластеризации, основанный на плотности. 

Метод не применим к данным с большой разницей в плотности кластеров.

Основные преимущества:
- Не требует априорного знания о необходимом количестве кластеров;
- Выделяет кластеры произвольной формы;
- Устойчив к выбросам, выделяет шум.

Подбор параметров

Вся сложность применения метода заключается в подборе двух основных параметров: min_samples и eps

1. min_samples параметр определяет сколько объектов должно содержаться в eps окрестности выбранного объекта, чтобы этот объект считалась внутрикластерным. Имеет смысл брать только значения ≥ 3, на практике часто бывает достаточно ограничить поиски значениями от 3 до 9, но это носит лишь рекомендательный характер и не отрицает наличия оптимальной пары с eps вне данного отрезка
2. eps параметр определяет размер окрестности. Чтобы больше определиться с областью поиска, для начала нужно выбрать функцию расстояния, так как для ненормированных векторов значения euclidean_distances и cosine_distances скорее всего лежат в разных диапазонах
3. Далее можно воспользоваться двумя методами подбора: 

- Метод колена: для выбранного min_samples определяется подходящий eps

Идея: для каждого объекта выборки считается расстояние до n-го соседа, где n выбирается равным min_samples . После этого расстояния сортируются и отрисовываются. В качестве eps берётся значение расстояния в точке с максимальной [кривизной](http://www.math24.ru/кривизна-плоских-линий.html), предполагая, что точки с сильно растущим расстоянием до n-го соседа являются выбросами.

Для поиска оптимальной точки можно воспользоваться библиотекой [kneed](https://kneed.readthedocs.io/en/stable/index.html) и ее классом [KneeLocator](https://kneed.readthedocs.io/en/stable/api.html#kneelocator). Суть его заключается в построении интерполирующей функции по заданным точкам (часто из класса полиномиальных функций) и выбора точки (среди заданных) с максимальным значением кривизны полученного интерполянта. В общепринятой формулировке метода колена значения расстояний располагаются в порядке возрастания, следовательно, имеем выпуклую кривую curve='convex' с порядком обхода по возрастанию direction='increasing', параметр interp_method='interp1d' or 'polynomial' отвечает методу интерполяции точек.

```python
from sklearn.neighbors import NearestNeighbors
from kneed import KneeLocator

n = 7

nearest_neighbors = NearestNeighbors(n_neighbors=n, metric='cosine', n_jobs=-1)
neighbors = nearest_neighbors.fit(embeddings)
distances, _ = neighbors.kneighbors(embeddings)
sort_distances = np.sort(distances[:, n - 1])

knee = KneeLocator(np.arange(len(sort_distances)), 
                   sort_distances, 
                   curve='convex', 
                   direction='increasing', 
                   interp_method='polynomial')

fig = plt.figure(figsize=(5, 5))
knee.plot_knee()
plt.xlabel("Points")
plt.ylabel("Distance")

print(distances_[knee.knee])
```

В качестве параметров для DBSCAN выбираются значения eps=0.11, min_samples=7, metric='cosine'.

На практике бывает полезно перебрать eps в небольшой окрестности найденного оптимального значения из-за зависимости оптимума от выбранного метода и степени интерполяции, попробовать разную интерполяцию и разные значения для n.

- Метод силуэта: 

Идея: перебирая по сетке значения для min_samples и eps , выбрать те, которые приводят к наибольшему значению [силуэт-метрики](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.silhouette_score.html).

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

for eps in np.linspace(0.07, 0.15, 10):
    for ms in [5, 7, 9]:
        clustering = DBSCAN(eps=eps, min_samples=ms, metric='cosine', n_jobs=-1)
        clusters = clustering.fit_predict(embeddings)
        score = silhouette_score(embeddings, clusters)
        print(f'eps = {eps:0.3f}, ms = {ms}: {score:0.4f}')
```

#### Задание: разбиение категории объявлений на виды товаров

Категория — это очень широкая группа товаров, часто объединённых общим предназначением. Например, в категории «Фитнес и тренажёры» находятся как беговая дорожка, так и фитнес-браслет. Это очень разные предметы по размеру и по виду, при этом они имеют верхнеуровневое общее предназначение — используются для занятий спортом.

Зачем же нам нужно разбиение на виды товаров? Рассмотрим примеры задач.


##### Первый пример: доставка

Представим, что покупателю понадобилась редкая запчасть или захотелось купить авторскую сумочку в единственном экземпляре. Что делать, если они находятся в другом городе? Для таких случаев нужна возможность доставки. Не все товары подлежат доставке — например, мы не можем доставить очень большие и тяжёлые или опасные товары. Но как понять, что мы можем отправлять, а что нет?

Нужна модель, которая будет это предсказывать. С одной стороны, мы могли бы определить признак доставляемости и таким образом разделить товары. Но что делать, если условия доставки изменились, и мы можем доставлять ещё какие-то товары?

Тогда нужно менять разметку, переобучать модель, хотя мы могли бы воспользоваться более универсальным решением: обучить модель определять вид товара, а уже после обрабатывать её решение фильтром доставляемости. Скорее всего, все беговые дорожки мы доставить не можем, в отличии от фитнес-браслетов. В таком случае, если у нас изменился допустимый к доставке список товаров, нам ничего не стоит обновить наш механизм — нужно только поменять фильтр.


##### Второй пример: введение тарифов

Например, нам бы хотелось ввести разные цены и лимиты на размещение и продвижение объявлений в зависимости от стоимости товара. В одной общей куче, где есть и дорогие, и дешёвые товары, это сделать сложно. И поэтому нужно как-то различать товары в категории. Для такой задачи также универсальным решением будет определять сначала вид товара, а уже потом применять к нему какие-то действия.

 
##### Третий пример: автоматическая модерация.

Схожее решение используется при автоматической проверке объявлений на допустимость к публикации, поскольку не все товары можно продавать на Авито (например, наркотики, оружие или животных из Красной книги). 

В общем случае для разбиения категории есть текстовые данные заголовков, картинки товаров из объявлений и описания объявлений. В текущем задании мы ограничимся первыми двумя: заголовками и заглавными картинками.

Обычно полученные кластера проверяются на соответствие требованиям, которые вы поставили перед кластеризацией. Требования определяются из личного представления о хорошем разбиении или формируются при анализе данных. Также на практике часто производится разметка небольшого количества данных для проведения валидации (при выполнении задания вы также можете разметить немного данных ~100 примеров, чтобы валидироваться).
